```python
import numpy as np
```

```python
np.random.seed(42)

X_test = np.random.rand(100, 4) * 10
true_w = np.array([3.5, -1.2, 2.0, -0.5])
true_b = 5.0

noise = np.random.randn(100) * 0.1
y_test = np.dot(X_test, true_w) + true_b + noise

w_init = np.zeros(4)
b_init = 0.0
```

```python
def predict(X, w, b):

    return np.dot(X, w) + b  # simply gets a prediction

def cost(y_hat, y):
    num = y.shape[0]
    cost = (y - y_hat) ** 2
    
    return np.sum(cost) / (2*num) # squared cost error

def gradients(X, y, y_hat):
    m = X.shape[0]

    dj_dw = X.T @ (y_hat - y) / m  # transposed the matrix to get the dot product 
    dj_db = np.sum(y_hat - y) / m
    

    return dj_dw, dj_db
```

```python
def gradient_descent(X, y, w, b, alpha, iterations):

    for i in range(iterations):

        y_hat = predict(X, w, b)

        dj_dw, dj_db = gradients(X, y, y_hat)

        w = w - (alpha * dj_dw)
        b = b - (alpha * dj_db)

        if i % 50 == 0:

            new_cost = cost(y_hat, y)
            print(f"Epoch {i}: Cost = {new_cost}")
            print(f"W: {w}, b: {b}")
    
    return w, b
```

```python
#Running time

gradient_descent(X_test, y_test, w_init, b_init, 0.0001, 1000)
```
