```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(42) # keeps track of y_train
y_train = rng.integers(1, 900, 100) # price in thousands
x_train = np.linspace(1, 5, 100) # in 1000 sqft 


def initialize_and_predict(X): ## simple model with 0.0 parameters

    w = 0.0
    b = 0.0

    y_hat = w * X + b
    return y_hat


print(initialize_and_predict(x_train).shape)
```

```python
def cost_function(y, y_hat):
    
    num = y.shape[0]

    cost = (y - y_hat) ** 2  ## broadcasting instead of looping will make the run time of my code be faster 
    cost = np.sum(cost)
    total_cost = cost / (2*num)

    return total_cost

cost_function(y_train, initialize_and_predict(x_train))
```

```python
def compute_gradients(X, y, y_hat):
    
    iterations = y.shape[0]

    dj_dw = (y_hat - y) * X
    dj_dw = np.sum(dj_dw) / (iterations)

    dj_db = (y_hat - y)
    dj_db = np.sum(dj_db) / (iterations)

    return dj_dw, dj_db
```

```python
def linear_regression(X, y, alpha, epochs):

    w = 0.0
    b = 0.0

    for epoch in range(epochs):
        
        y_hat = w * X + b

        dj_dw, dj_db = compute_gradients(X, y, y_hat)

        w = w - (alpha * dj_dw)
        b = b - (alpha * dj_db)

        if epoch % 100 == 0:
            new_y_hat = w * X + b
            
            current_cost = cost_function(y, new_y_hat)
            print(f"Epoch {epoch}: Cost = {current_cost}")
    
    return w, b
```

```python
linear_regression(x_train, y_train, 0.1, 1000)
```
